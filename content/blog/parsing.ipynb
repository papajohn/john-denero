Sentences have hierarchical structure. Understanding the intended meaning of a sentence often requires determining that structure from its ordered words. For example, consider this famous sentence from the preface to Structure and Interpretation of Computer Programs:

> Programs must be written for people to read

The tree below corresponds to the intended meaning: that *computer programs should be composed with people as the intended audience*.

```python
s=("Sentence", 
   [("Noun Phrase", 
     ["Programs"]),
    ("Verb Phrase", 
     ["must be written",
      ("Subordinate Clause", 
       ["for people to read"])])])
from draw_tree import draw_tree
draw_tree(s)
```

In Natural Language Processing (NLP), this tree is called a *phrase-structure parse tree* of the sentence. Spans of the original sentence are grouped hierarchically into *constituents*, and each constituent has a *tag* such as *noun phrase* or *sentence*.

Sentences in English are often ambiguous. A second possible interpretation of this sentence is that *programs must first be written in order for people to read them*.  In this case, the subordinate clause is not part of the verb phrase, but instead modifies the whole sentence.

```python
s=("Sentence", 
   [("Noun Phrase", ["Programs"]),
    ("Verb Phrase", ["must be written"]),
    ("Subordinate Clause", ["for people to read"])])
draw_tree(s)
```

A third (somewhat obscure) interpretation is that *programs must be written for people, in order for those programs to be able to read*. In this case, "for people to read" is not a single constituent.

```python
s=("Sentence", 
   [("Noun Phrase", ["Programs"]),
    ("Verb Phrase", ["must be written",
                     ("Prepositional\nPhrase", ["for people"])]),
    ("Infinitive", ["to read"])])
draw_tree(s)
```

A *parser* is a program that infers plausible trees for a sentence. Statistical parsers, which are used heavily in NLP systems, not only enumerate alternatives but also score likely trees above obscure ones. In this example, a good statistical parser should infer that:

* "programs" is not a verb (as in *she programs for fun*) but instead a plural noun, and
* "must" is not a noun (as in *dampness or mold*) but instead a modal  verb,
* "read" is not a noun (as in *what a great read*) but an infinitival verb, etc.

The remainder of this post defines a basic statistical parser by learning the common structural patterns of the English language from examples.

## Representing Trees

To build a parser, we first need data types to represent parse trees. A ``Tree`` has a ``tag`` and a sequence of ``branches`` that can be either ``Tree`` instances or ``Leaf`` instances. Each ``Leaf`` is a single word.

```python
class Tree:
    """A phrase-structure parse tree."""
    def __init__(self, tag, branches):
        assert len(branches) >= 1
        for b in branches:
            assert isinstance(b, Tree) or isinstance(b, Leaf)
        self.tag = tag
        self.branches = branches

class Leaf:
    """A word at the leaf of a Tree."""
    def __init__(self, tag, word):
        self.tag = tag
        self.word = word
```

We can create some interesting sentence structures using a single word, *buffalo*. Buffalo ([plural noun](http://en.wiktionary.org/wiki/buffalo#Noun)) are large mammals, and *to buffalo* ([verb](http://en.wiktionary.org/wiki/buffalo#Verb)) is to intimidate. As herd animals, buffalo tend to intimidate other buffalo.

```python
beasts = Leaf('N', 'buffalo')
intimidate = Leaf('V', 'buffalo')

s = Tree('S', [Tree('NP', [beasts]),
               Tree('VP', [intimidate,
                           Tree('NP', [beasts])])])
draw_tree(s)
```

In English, "that" can be used to begin a *relative clause* for which we use the tag **RP**. The word "that" is optional, and below we omit it.

```python
s = Tree('S', [Tree('NP', [beasts, # (that)
                           Tree('RP', [Tree('NP', [beasts]),
                                       Tree('VP', [intimidate])])]),
               Tree('VP', [intimidate,
                           Tree('NP', [beasts])])])
draw_tree(s)
```

### Context-Free Grammars

A context-free grammar is a set of *rules* that compactly describe the sentences in a language. For simplicity, we can start with a grammar for the buffalo language: valid English sentences that only contain the word "buffalo" repeated. Conventionally, rules are written with a single symbol on the left expanding to a sequence of symbols on the right. For example,

> S $\rightarrow$ NP VP

> NP $\rightarrow$ N

> N $\rightarrow$ buffalo

> VP $\rightarrow$ V NP

> V $\rightarrow$ buffalo

These five rules form a grammar that allows the sentence symbol ``S`` to expand to the sentence "buffalo buffalo buffalo." The ``S`` expands to ``NP`` followed by ``VP`` by the first rule. That ``NP`` expands to ``N`` which in turn expands to ``buffalo``. That ``VP`` expands recursively to "buffalo buffalo" through repeated expansion. The parse tree for this sentence shown above is a trace of these expansions.

To implement this grammar, we can represent these rules using built-in Python data structures. Here, we separate the rules that expand to words (the ``lexicon``) and the rules that expand to more symbols (the ``grammar``). Formally, they are all part of the same grammar.

```python
lexicon = {
    Leaf('N', 'buffalo'), # mammals
    Leaf('V', 'buffalo'), # intimidate
    }

grammar = {
    'S':  [['NP', 'VP']],
    'NP': [['N']],
    'VP': [['V', 'NP']],
    }
```

The ``expand`` function is a recursive function that takes as input a single tag and recursively expands that tag using the grammar and lexicon. 

```python
def expand(tag):
    """Yield all trees rooted by tag."""
    for leaf in lexicon:
        if leaf.tag == tag:
            yield leaf
    if tag in grammar:
        for tags in grammar[tag]:
            for branches in expand_all(tags):
                yield Tree(tag, branches)

def expand_all(tags):
    """Yield all sequences of branches for a sequence of tags."""
    if len(tags) == 1:
        for branch in expand(tags[0]):
            yield [branch]
    else:
        first, rest = tags[0], tags[1:]
        for first_branch in expand(first):
            for rest_branches in expand_all(rest):
                yield [first_branch] + rest_branches
```

Recursive expansion lets us inspect what sentences are generated by a grammar. It will also serve as a central part of the parsing algorithm we are developing. The algorithm is entirely brute-force; for each rule that has ``tag`` on its left-hand side, the ``expand`` function yields a tree for every way of expanding the tags on its right-hand side.

Printing the result of expanding ``S`` displays all sentences generated by our grammar. So far, only one parse tree can be constructed.

```python
def draw_trees(ts):
    for i, t in enumerate(ts):
        figure(i)
        draw_tree(t)
draw_trees(expand('S'))
```

The word "buffalo" can also be used as an adjective, describing something from Buffalo, New York. 

> NP $\rightarrow$ J N

> J $\rightarrow$ J

After adding these rules to our grammar, ``expand`` yields four results because each noun phrase (``NP``) may or may not have an adjective (``J``) modifying its head noun (``N``).

```python
lexicon.add(Leaf('J', 'buffalo')) # from New York
grammar['NP'].append(['J', 'N'])
draw_trees(expand('S'))
```

We can add relative clauses as well by adding the following rules.

> NP $\rightarrow$ NP RP

> RP $\rightarrow$ NP V

These rules state that any noun phrase (e.g., "mammals") can be modified by a relative clause that contains a subject noun phrase followed by a verb. Each noun phrase can contain a noun phrase, and so an infinite variety of sentences can be generated. Below, we show the first five.

```python
grammar['NP'].append(['NP', 'RP'])
grammar['RP'] = [['NP', 'V']]
from itertools import islice
draw_trees(islice(expand('S'), 0, 5))
```

### Parsing

Parsing is the problem of finding trees that yield a particular sequence of words. Given an input sequence of five "buffalo", we would like to enumerate all different syntactic interpretations of those words. 

One approach to parsing is to recursively expand ``S``, but enforce that the leaves of the resulting trees match the input words. To do so, we must keep track of the part of the sentence to which each subtree corresponds.

![parsing animation](parsing_animation.gif)

The ``parse`` function below is extremely similar to the ``expand`` function we already created.

```python
def parse(line):
    words = line.split()

    def expand(start, end, tag):
        """Yield all trees rooted by tag over words[start:end]."""
        if end-start == 1:
            word = words[start]
            for leaf in lexicon:
                if leaf.tag == tag and leaf.word == word:
                    yield leaf
        if tag in grammar:
            for tags in grammar[tag]:
                for branches in expand_all(start, end, tags):
                    yield Tree(tag, branches)

    def expand_all(start, end, tags):
        """Yield all sequences of branches for tags over words[start:end]."""
        if len(tags) == 1:
            for branch in expand(start, end, tags[0]):
                yield [branch]
        else:
            first, rest = tags[0], tags[1:]
            for middle in range(start+1, end+1-len(rest)):
                for first_branch in expand(start, middle, first):
                    for rest_branches in expand_all(middle, end, rest):
                        yield [first_branch] + rest_branches

    for tree in expand(0, len(words), 'S'):
        yield tree
```

```python
# All possible parses of five buffalo!
draw_trees(parse('buffalo buffalo buffalo buffalo buffalo'))
```
