## Natural Language Parsing

Sentences have hierarchical structure. Understanding the intended meaning of a sentence often requires determining that structure from its ordered words. For example, consider this famous sentence from the preface to Structure and Interpretation of Computer Programs:

> Programs must be written for people to read

The tree below corresponds to the intended meaning: that *computer programs should be composed with people as the intended audience*.

```python
s=("Sentence", 
   [("Noun Phrase", 
     ["Programs"]),
    ("Verb Phrase", 
     ["must be written",
      ("Subordinate Clause", 
       ["for people to read"])])])
from draw_tree import draw_tree
draw_tree(s)
```

In Natural Language Processing (NLP), this tree is called a *phrase-structure parse tree* of the sentence. Spans of the original sentence are grouped hierarchically into *constituents*, and each constituent has a *tag* such as *noun phrase* or *sentence*.

Sentences in English are often ambiguous. A second possible interpretation of this sentence is that *programs must first be written in order for people to read them*.  In this case, the subordinate clause is not part of the verb phrase, but instead modifies the whole sentence.

```python
s=("Sentence", 
   [("Noun Phrase", ["Programs"]),
    ("Verb Phrase", ["must be written"]),
    ("Subordinate Clause", ["for people to read"])])
draw_tree(s)
```

A third (somewhat obscure) interpretation is that *programs must be written for people, in order for those programs to be able to read*. In this case, "for people to read" is not a single constituent.

```python
s=("Sentence", 
   [("Noun Phrase", ["Programs"]),
    ("Verb Phrase", ["must be written",
                     ("Prepositional\nPhrase", ["for people"])]),
    ("Infinitive", ["to read"])])
draw_tree(s)
```

A *parser* is a program that infers plausible trees for a sentence. Statistical parsers, which are used heavily in NLP systems, not only enumerate alternatives but also score likely trees above obscure ones. In this example, a good statistical parser should infer that:

* "programs" is not a verb (as in *she programs for fun*) but instead a plural noun, and
* "must" is not a noun (as in *dampness or mold*) but instead a modal  verb,
* "read" is not a noun (as in *what a great read*) but an infinitival verb, etc.

The remainder of this post defines a basic statistical parser by learning the common structural patterns of the English language from examples.

## Representing Trees

To build a parser, we first need data types to represent parse trees. A ``Tree`` has a ``tag`` and a sequence of ``branches`` that can be either ``Tree`` instances or ``Leaf`` instances. Each ``Leaf`` is a single word.

```python
class Tree:
    """A phrase-structure parse tree."""
    def __init__(self, tag, branches):
        assert len(branches) >= 1
        for b in branches:
            assert isinstance(b, Tree) or isinstance(b, Leaf)
        self.tag = tag
        self.branches = branches

class Leaf:
    """A word at the leaf of a Tree."""
    def __init__(self, tag, word):
        self.tag = tag
        self.word = word
```

We can create some interesting sentence structures using a single word, *buffalo*. Buffalo ([plural noun](http://en.wiktionary.org/wiki/buffalo#Noun)) are large mammals, and *to buffalo* ([verb](http://en.wiktionary.org/wiki/buffalo#Verb) is to intimidate. As social herd animals, buffalo tend to intimidate other buffalo.

```python
mammals = Leaf('N', 'buffalo')
intimidate = Leaf('V', 'buffalo')

s = Tree('S', [Tree('NP', [mammals]),
               Tree('VP', [intimidate,
                           Tree('NP', [mammals])])])
draw_tree(s)
```

Once some buffalo are buffaloed by other buffalo, they feel the need to buffalo some buffalo themselves. That is, the buffalo that buffalo buffalo sometimes buffalo other buffalo. In English, "that" can be used to begin a *relative clause* for which we use the tag **RP**. The word "that" is optional, and below we omit it.

```python
s = Tree('S', [Tree('NP', [mammals, # (that)
                           Tree('RP', [Tree('NP', [mammals]),
                                       Tree('VP', [intimidate])])]),
               Tree('VP', [intimidate,
                           Tree('NP', [mammals])])])
draw_tree(s)
```
